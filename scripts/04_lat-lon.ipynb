```python
import pandas as pd
import urllib.request
import json
import io
```

```python
meta_data = pd.read_csv("../data/raw_shapes/ESP_meta.csv")[["STATION", "NAME", "SQ_KM"]]
#consistency
meta_data.at[314, "STATION"] = 38001
meta_data.at[315, "STATION"] = 39001
```

```python
ids  = []
lats = []
lons = []
for idx in meta_data["STATION"].values:
    
    idx = str(int(idx))
    
    print(f"{idx}")
    
    request = f"http://nrfaapps.ceh.ac.uk/nrfa/ws/station-info/{idx}?station={idx}&format=json-object&fields=lat-long"
    
    data = urllib.request.urlopen(request).read()
    
    data = json.loads(data)
    
    lat = data["data"][0]["lat-long"]["latitude"]
    
    lon = data["data"][0]["lat-long"]["longitude"]
    
    ids.append(idx)
    lats.append(lat)
    lons.append(lon)
```

```python
ll = pd.DataFrame({"station_id": ids, "lat": lats, "lon": lons})
```

```python
ll.to_pickle("../data/raw_shapes/ESP_latlon.pkl")
```
