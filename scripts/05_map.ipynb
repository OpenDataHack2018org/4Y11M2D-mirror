```python
# source activate folium
import folium
import branca
from folium.features import CustomIcon
from folium.plugins import FloatImage
import pandas as pd
```

```python
meta = pd.read_pickle("../data/raw_shapes/ESP_latlon.pkl")
```

```python
m = folium.Map([54, -3], zoom_start=6)

for i, row in meta.iterrows():
    
    idx = row["station_id"]
    lat = row["lat"]
    lon = row["lon"]
    
    html = f"\n<iframe src='https://hydrogo.github.io/4Y11M2D/{idx}.html' scrolling='no' frameborder='no' width=600 height=360></iframe>\n"
    
    iframe = branca.element.IFrame(html=html, width=620, height=380)
    popup = folium.Popup(iframe, max_width=550)
    folium.Marker([lat, lon], popup=popup, icon=folium.Icon(icon="tint", color="darkblue")).add_to(m)

FloatImage("https://raw.githubusercontent.com/hydrogo/4Y11M2D/master/docs/4Y11M2D_logo_small.png", 
           bottom=2, left=2).add_to(m)
```
