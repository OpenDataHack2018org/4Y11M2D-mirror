```python
import geojson
import geopandas as gpd
from shapely.geometry import Polygon as shpPolygon

%matplotlib inline
```

```python
shapes = gpd.read_file("../data/raw_shapes/ESP_updatedCatchList_Nov2016_ALL_mod_314.shp")

shapes.crs = {'proj': 'tmerc',
 'lat_0': 49,
 'lon_0': -2,
 'k': 0.9996012717,
 'x_0': 400000,
 'y_0': -100000,
 'datum': 'OSGB36',
 'units': 'm',
 'no_defs': True}

shapes = shapes.to_crs(epsg=4326)
```

```python
# print(shapes.head())

shapes.plot()
```

```python
for i in shapes["allCatch_I"]:
    
    part = shapes[shapes["allCatch_I"] == i]
    
    part.to_file(f"../data/shapes/{i}.geojson", driver="GeoJSON")
```
